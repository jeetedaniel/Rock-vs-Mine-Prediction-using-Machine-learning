Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
import lightgbm as lgb
```

Data Collection and Data Processing

```python
#Loading a dataset into a Pandas DataFrame
sonar_data = pd.read_csv('/content/Copy of sonar data.csv',header=None)
```

```python
sonar_data.head()
```

```python
#number of rows and colums
sonar_data.shape
```

```python
#Showing Statistical Defenitions of the Dataset
sonar_data.describe()
```

```python
#Determining the number of different Catagorical Value
sonar_data[60].value_counts()
```

M-->Mines
R-->Rocks

```python
#Finding Mean for Dataset Grouped by Mine or Rocks
sonar_data.groupby(60).mean()
```

```python
#Seperating Data and the Labels
X = sonar_data.drop(columns=60,axis=1)
Y = sonar_data[60]
```

```python
#Printing the seperated Dataset
print(X)
print(Y)
```

Training and Test Data splitting



```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training --> Logistic Regression Model (Good for Binary Classification Labels)


```python
#Creating a Logistic Regression Model
model = LogisticRegression()

#Creating a SVM model
classifier = svm.SVC(kernel='linear')

#Creating a Naive Bayes Model
gnb = GaussianNB()

#creating a Random Forest Model
rf = RandomForestClassifier(n_estimators=40)

#Creating LGBMC model
clf = lgb.LGBMClassifier()
```

```python
#training the models with training dataset
model.fit(X_train,Y_train)
classifier.fit(X_train,Y_train)
gnb.fit(X_train,Y_train)
rf.fit(X_train,Y_train)
clf.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy on training Data using Logistic Regression
prediction_logistic = model.predict(X_train)
training_data_accuracy = accuracy_score(prediction_logistic,Y_train)

#Accuracy on Train Data using SVM
prediction_svm = classifier.predict(X_train)
training_data_accuracy_svm = accuracy_score(prediction_svm,Y_train)

#Accuracy on Train Data using Naive_bayes
prediction_gnb = gnb.predict(X_train)
X_train_prediction_gnb = accuracy_score(prediction_gnb,Y_train)

#Accuracy on Train Data using RandomForest
prediction_rf = rf.predict(X_train)
X_train_prediction_rf = accuracy_score(prediction_rf,Y_train)

#Accuracy on Train Data using LGBMClassifier
prediction_lgbm = clf.predict(X_train)
X_train_prediction_lgbm = accuracy_score(prediction_lgbm,Y_train)
```

```python
#printing Accuracy Score using Logistic Regression
print("Model Accuracy on Training Data using Logistic Regression: ",training_data_accuracy)

#printing Accuracy score using SVM
print("Model Accuracy on Training Data using SVM : ",training_data_accuracy_svm)

#printing Accuracy Score using Naive Bayes
print("Model Accuracy on Training Data using Naive Bayes : ",X_train_prediction_gnb)

#printing Accuracy score using Random Forest
print("Model Accuracy on Training Data using Random Forest : ",X_train_prediction_rf)

#Printing Accuracy score using LGBMClassifier
print("Model Accuracy on Training Data using LGBMClassifier : ",X_train_prediction_lgbm)
```

```python
#Accuracy Score on Test Data using Logistic Regression
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)

#Accuracy score on Test Data using SVM
Prediction_svm = classifier.predict(X_test)
test_accuracy_svm =accuracy_score(Prediction_svm,Y_test)

#Accuracy score on Test Data using Naive Bayes
prediction_gnb = gnb.predict(X_test)
X_test_prediction_gnb= accuracy_score(prediction_gnb,Y_test)

#Accuracy score on Test Data using Random Forest
prediction_rf = rf.predict(X_test)
X_test_prediction_rf = accuracy_score(prediction_rf,Y_test)

#Accuracy on Test Data using LGBMClassifier
prediction_lgbm = clf.predict(X_test)
X_test_prediction_lgbm = accuracy_score(prediction_lgbm,Y_test)
```

```python
#printing Accuracy Score using Logistic Regression
print("Model Accuracy on Test Data using Logistic Regression : ",test_data_accuracy)

#printing Accuracy Score using SVM
print("Model Accuracy on Test Data using SVM : ",test_accuracy_svm)

#Printing Accuracy Score using Naive Bayes
print("Model Accuracy on Test Data using Naive Bayes : ",X_test_prediction_gnb)

#Printing Accuracy Score using Random Forest
print("Model Accuracy on Test Data using Random Forest : ",X_test_prediction_rf)

#Printing Accuracy score using LGBMClassifier
print("Model Accuracy on Test Data using LGBMClassifier : ",X_test_prediction_lgbm)
```
